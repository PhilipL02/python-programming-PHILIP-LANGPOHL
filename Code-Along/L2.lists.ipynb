```python
ls = [1, 2, 3, 4, 5]
new_list = [l for l in ls if l % 2 == 0]
print(new_list)

new_list.append(2)

print(new_list)
```

```python
lst = {i for i in range(1, 11)}
print(lst)

lst.add(1)

print(lst)
```

```python
import matplotlib.pyplot as plt
x = list(range(10))
k = 2
m = 2
y = [(k * x + m) for x in x]

plt.plot(x, y)
plt.title(f'Plot y = {k}x + {m}')
plt.xlabel("x")
plt.ylabel("y")
```

### Regular expressions

```python
import re

text = "Det var en gång en katt som tog på sig en hatt."

match_list = re.findall(r".att", text)

print(match_list)
```

```python
person_numbers = "Kalle: 19990124-0192, Bengt: 19680420-1020, Inger: 19550703-9201, Pontus: 010410-2010"

match_long = re.findall(r"\d{8}-\d{4}", person_numbers)

print(f"YYYYMMDD-XX: {match_long}")
```

```python
match_short = re.findall(r"\d{6}-\d{4}", person_numbers)

print(f"YYMMDD-XX: {match_short}")
```

```python
my_match = re.findall(r"\d+-\d+", person_numbers)

print(my_match)
```
