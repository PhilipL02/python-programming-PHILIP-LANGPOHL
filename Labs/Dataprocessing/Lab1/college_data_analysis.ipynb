## Exploring and Analyzing Data of US Universities and Colleges

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

college = pd.read_csv("College.csv")
```

```python
college
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")
```

```python
college3
```

```python
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
plt.show()
```

```python
college.boxplot(column="Outstate", by='Private')
plt.show()
```

```python
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])
college["Elite"].value_counts()
```

```python
college.boxplot(column="Outstate", by='Elite')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

plot_configurations = {
    "Expend": {
        "bins": 30,
        "title": "Instructional expenditure per student",
        "color": "skyblue"
    },
    "S.F.Ratio": {
        "bins": 20,
        "title": "Student/faculty ratio",
        "color": "lightgreen"
    },
    "Outstate": {
        "bins": 12,
        "title": "Out-of-state tuition",
        "color": "salmon"
    },
    "Grad.Rate": {
        "bins": 16,
        "title": "Graduation rate",
        "color": "lightcoral"
    }
}

for ax, (column_name, config) in zip(axes.flatten(), plot_configurations.items()):
    college[column_name].plot.hist(
        bins=config["bins"],
        ax=ax,
        color=config["color"],
        edgecolor="black"
    )
    ax.set_title(config["title"])
    ax.set_xlabel(column_name)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
college['Accept_Ratio'] = college['Accept'] / college['Apps']
college['Accept_Ratio'].plot.hist(bins=20, color='lightblue', edgecolor='black')
plt.show()
```

Jag tyckte det kunde vara intressant att kolla ungefär hur stor andel av de sökande som faktiskt blev accepterade.

Med detta histogram kan vi se att många skolor accepterar runt 80% av deras ansökningar.

```python
college.plot.scatter(x='Expend', y='S.F.Ratio')
plt.show()
```

Här väljer jag att kolla om det finns koppling mellan S.F.Ratio (Förhållande mellan antal studenter och lärare) och Expend (Undervisningskostnad per student).

Jag tänker att om en skola har ett få antal studenter per lärare borde det kunna bli en högre undervisningskostnad per student.

Detta tycker jag ändå att grafen visar. De skolor som har högst Expend ligger relativt lågt i S.F.Ratio, och de skolor med höga S.F.Ratio har ett relativt lågt värde i Expend.

```python
sns.scatterplot(x='Expend', y='S.F.Ratio', hue='Private', data=college)
plt.show()
```

Jag valde nu även att få med variabeln Private i grafen för att se om värdena skiljer sig åt om det är privat eller offentlig skola.

Man kan se att de offentliga skolorna för det mesta verkar ha ett högre förhållande mellan antal studenter och lärare än vad de privata har, även om det är några privata skolor som har högt värde där.

Man kan också se att de skolor som sticker ut i Expend är privata skolor.

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

college.boxplot(column="Grad.Rate", by='Private', ax=axes[0])
axes[0].set_title("Graduation Rate by Private/Public")
axes[0].set_xlabel("Private")
axes[0].set_ylabel("Graduation Rate (%)")

college.boxplot(column="Grad.Rate", by='Elite', ax=axes[1])
axes[1].set_title("Graduation Rate by Elite Status")
axes[1].set_xlabel("Elite")
axes[1].set_ylabel("Graduation Rate (%)")

fig.suptitle("Comparison of Graduation Rate by Institution Type", fontsize=14)
plt.tight_layout()
plt.show()
```

Här ovan har jag valt att undersöka Grad.Rate (Procentandel av studenterna som tar examen), och se hur det skiljer sig åt beroende på Private och Elite.

Som det ser ut är det privata skolor som generellt har bättre Grad.Rate, även om det finns en del privata skolor med dålig rate också.

Det är ännu större skillnad när man jämför Elite status. Alla skolor som är klassade som Elite verkar ha högre än 50% i Grad.Rate, medan de som inte är Elite kan ha betydligt lägre.

Detta känns rimligt eftersom de som klassas som Elite är de skolor som fått in en stor andel studenter som kom från topp 10% i sitt gymnasium. Det känns då troligt att dessa elever kommer lyckas ta examen.

```python
sns.scatterplot(x='Grad.Rate', y='Expend', hue='Elite', data=college)
plt.show()
```

Jag valde att gå vidare med detta att Elite-skolor ofta har en högre Grad.Rate, och valde att visualisera detta ihop med Expend (Undervisningskostnad per student).

Även denna visar då tydligt att alla Elite ligger långt till höger (med en hög Grad.Rate), men man kan också de att det är skolorna som blev klassade som Elite som är de som sticker ut mest i Expend.

Dock kan vi se en skola som inte är klassad som Elite, som har en mycket hög undervisningskostnad per student. Denna skolan ligger inte heller så bra i Grad.Rate. Detta borde nog denna skola kolla på. Den sticker ut rejält i Expend i förhållande till skolorna med liknande Grad.Rate.
