## Exploring and Analyzing Data of US Universities and Colleges

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

college = pd.read_csv("College.csv")
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
plt.show()
```

```python
college.boxplot(column="Outstate", by='Private')
plt.show()
```

```python
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])
college["Elite"].value_counts()
```

```python
college.boxplot(column="Outstate", by='Elite')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

plot_configurations = {
    "Expend": {
        "bins": 30,
        "title": "Instructional expenditure per student",
        "color": "skyblue"
    },
    "S.F.Ratio": {
        "bins": 20,
        "title": "Student/faculty ratio",
        "color": "lightgreen"
    },
    "Outstate": {
        "bins": 12,
        "title": "Out-of-state tuition",
        "color": "salmon"
    },
    "Grad.Rate": {
        "bins": 16,
        "title": "Graduation rate",
        "color": "lightcoral"
    }
}

for ax, (column_name, config) in zip(axes.flatten(), plot_configurations.items()):
    college[column_name].plot.hist(
        bins=config["bins"],
        ax=ax,
        color=config["color"],
        edgecolor="black"
    )
    ax.set_title(config["title"])
    ax.set_xlabel(column_name)
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
college['Accept_Ratio'] = college['Accept'] / college['Apps']
college['Accept_Ratio'].plot.hist(bins=20, color='lightblue', edgecolor='black')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

college.boxplot(column="Grad.Rate", by='Private', ax=axes[0])
axes[0].set_title("Graduation Rate by Private/Public")
axes[0].set_xlabel("Private")
axes[0].set_ylabel("Graduation Rate (%)")

college.boxplot(column="Grad.Rate", by='Elite', ax=axes[1])
axes[1].set_title("Graduation Rate by Elite Status")
axes[1].set_xlabel("Elite")
axes[1].set_ylabel("Graduation Rate (%)")

fig.suptitle("Comparison of Graduation Rate by Institution Type", fontsize=14)
plt.tight_layout()
plt.show()
```
