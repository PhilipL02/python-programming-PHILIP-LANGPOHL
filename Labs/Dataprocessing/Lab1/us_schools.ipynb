```python
# Skapa ett 2x2 subplotsfÃ¶nster
fig, axes = plt.subplots(2, 2, figsize=(10, 8))  # storlek pÃ¥ figuren kan anpassas

# Rita histogram med olika antal bins
college['Apps'].plot.hist(bins=40, ax=axes[0, 0], color='skyblue', edgecolor='black')
axes[0, 0].set_title('Applications Received (Apps)')

college['Accept'].plot.hist(bins=15, ax=axes[0, 1], color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Applicants Accepted (Accept)')

college['Outstate'].plot.hist(bins=10, ax=axes[1, 0], color='salmon', edgecolor='black')
axes[1, 0].set_title('Out-of-state Tuition (Outstate)')

college['Grad.Rate'].plot.hist(bins=12, ax=axes[1, 1], color='lightcoral', edgecolor='black')
axes[1, 1].set_title('Graduation Rate (Grad.Rate)')

# Justera layouten sÃ¥ att titlar och etiketter inte Ã¶verlappar
plt.tight_layout()
plt.show()
```

```python
college4 = college
college4['Accept_Ratio'] = college4['Accept'] / college4['Apps']
average_ratio = college4['Accept_Ratio'].mean()
print(f"Genomsnittligt acceptansfÃ¶rhÃ¥llande: {average_ratio:.2f}")
college['Accept_Ratio'].plot(kind='hist', bins=20, color='lightblue', edgecolor='black')
```

```python
fig, ax = plt.figure(dpi=100),plt.axes()
college.plot.scatter(x='Apps', y='Accept', c='DarkBlue', ax=ax)
x = np.linspace(min(college["Apps"]), max(college["Apps"]))
y = [x*average_ratio for x in x]
plt.plot(x, y)
plt.show()
```
