## Exploring data of US universities and colleges

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

college = pd.read_csv("College.csv")
```

```python
college
```

```python
college2 = pd.read_csv("College.csv", index_col=0)
college3 = college.rename({"Unnamed: 0": "College"}, axis=1)
college3 = college3.set_index("College")
```

```python
college3
```

```python
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
college.boxplot(by='Private', column="Outstate")
```

```python
college["Elite"] = pd.cut(college["Top10perc"], [0, 50, 100], labels=["No", "Yes"])
college["Elite"].value_counts()
```

```python
college.boxplot(by='Elite', column="Outstate")
```

Vi kan se att värdet på Outstate generellt är större för skolor som är "Elite".

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

plot_settings = {
    "Apps": {
        "bins": 25,
        "title": "Number of applications received",
        "color": "skyblue"
    },
    "Accept": {
        "bins": 30,
        "title": "Number of applicants accepted",
        "color": "lightgreen"
    },
    "Outstate": {
        "bins": 20,
        "title": "Out-of-state tuition",
        "color": "salmon"
    },
    "Grad.Rate": {
        "bins": 10,
        "title": "Graduation rate",
        "color": "lightcoral"
    }
}

for ax, (key, attributes) in zip(axes.flatten(), plot_settings.items()):
    college[key].plot.hist(bins=attributes["bins"], ax=ax, color=attributes["color"], edgecolor='black')
    ax.set_title(attributes["title"])

plt.tight_layout()
plt.show()
```

```python
college4 = college
college4['Accept_Ratio'] = college4['Accept'] / college4['Apps']
average_ratio = college4['Accept_Ratio'].mean()
print(f"Genomsnittligt acceptansförhållande: {average_ratio:.2f}")
college['Accept_Ratio'].plot(kind='hist', bins=20, color='lightblue', edgecolor='black')
```

```python
fig, ax = plt.figure(dpi=100),plt.axes()
college.plot.scatter(x='Apps', y='Accept', c='DarkBlue', ax=ax)
x = np.linspace(min(college["Apps"]), max(college["Apps"]))
y = [x*average_ratio for x in x]
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.barplot(x='Private', y='Grad.Rate', data=college, estimator=np.mean)
plt.title("Average Graduation Rate by Institution Type")
plt.show()
```

```python
sns.barplot(x='Elite', y='Grad.Rate', data=college, estimator=np.mean)
plt.title("Average Graduation Rate by Elite status")
plt.show()
```

Om skolan klassas som Elite är genomsnittliga Grad.Rate högre än om skolan inte är Elite, vilket känns rimligt och troligt.

```python
sns.barplot(x='Elite', y='Personal', data=college, estimator=np.mean)
plt.title("Average estimated personal spending by Elite status")
plt.show()
```

Skolor som är klassade som Elite spenderar i genomsnitt mindre pengar på personal.

```python
sns.scatterplot(x='Outstate', y='F.Undergrad', hue='Private', data=college)
plt.title("Out-of-State Tuition vs. Full-Time Undergraduates")
plt.show()
```

```python
sns.pairplot(college[['Outstate', 'Room.Board', 'Grad.Rate', 'perc.alumni', "Private"]], hue='Private')
plt.show()
```

```python
sns.violinplot(x='Private', y='Grad.Rate', data=college)
plt.title("Distribution of Graduation Rates by Institution Type")
plt.show()
```

```python
college.groupby('Private')[['Top10perc', 'Top25perc']].mean().plot(kind='bar', stacked=True)
plt.title("Proportion of Top Students by Institution Type")
plt.ylabel("Average Percentage")
plt.show()
```

```python
sns.lmplot(x='Expend', y='Grad.Rate', hue='Private', data=college)
plt.title("Instructional Expenditure vs. Graduation Rate")
plt.show()
```

Högre kostnad per elev ser ut att resultera i bättre Grad.Rate. Alltså att större andel tar examen.

Vi kan även se att de skolor med högst kostnad per elev är privata skolor.

```python
from math import pi
categories = ['PhD', 'Terminal', 'S.F.Ratio', 'perc.alumni']
values = college[college['Private'] == 'Yes'][categories].mean().tolist()
values += values[:1]  # Circular data for radar plot

plt.figure(figsize=(6,6))
ax = plt.subplot(111, polar=True)
plt.xticks([n * 2 * pi / len(categories) for n in range(len(categories))], categories)
ax.plot(values)
ax.fill(values, alpha=0.3)
plt.title("Average Faculty & Student Characteristics for Private Colleges")
plt.show()
```
